```python
# Identify duplicate rows
duplicate_rows = df[df.duplicated()]

# Display duplicate rows
print("Duplicate Rows except first occurrence:")
print(duplicate_rows)
```

```python
# Drop duplicate rows
df_no_duplicates = df.drop_duplicates()

# Display DataFrame without duplicates
print("DataFrame without duplicates:")
print(df_no_duplicates)
```

```python

# Assuming df_no_duplicates is your DataFrame without duplicate rows
# If not, replace it with your actual DataFrame

# Download NLTK resources
nltk.download('stopwords')
nltk.download('punkt')

# Function for text cleaning
def clean_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))
    
    # Tokenize the text
    tokens = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    
    # Join tokens back into a cleaned text
    cleaned_text = ' '.join(tokens)
    
    return cleaned_text

# Apply text cleaning to the 'Comment' column
df_no_duplicates['Cleaned Comment'] = df_no_duplicates['Comment'].apply(clean_text)

# Display the DataFrame with the cleaned text
print(df_no_duplicates[['Comment', 'Cleaned Comment']])
```
